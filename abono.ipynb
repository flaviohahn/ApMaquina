As informações disponíveis nesta consulta referem-se aos servidores aposentados do Poder Executivo Federal, não incluindo portanto, dados sobre servidores ativos, pensionistas ou instituidores de pensão, nem mesmo sobre servidores dos poderes Judiciário e Legislativo. 

Os dados sobre servidores civis têm origem no Sistema Integrado de Administração de Recursos Humanos (Siape) e reúnem informações dos órgãos integrantes do Sistema de Pessoal Civil da Administração Federal (Sipec), e não incluem dados relativos aos servidores do Banco Central do Brasil (Bacen).

Os arquivos são os disponibilizados pelo Portal de dados abertos, e portanto, não são da fonte primária. Estão disponíveis informações, em formato aberto, sobre nome do aposentado, cpf (desconfigurado) órgão de origem, cargo emprego, regime jurídico, fundamentação da aposentadoria dos servidores públicos do executivo federal. *   Servidores: NOME (60); 

*   CPF(11)(descaracterizado);
 
*   MATRICULA(ORGAO(5) + MATRICULA(7));
  
*   NOME ÓRGÃO DE ORIGEM (40); 

*   SIGLA ÓRGÃO DE ORIGEM (10) ÓRGÃO VINCULADO (5); 

*   DESCRIÇÃO-CARGO-EMPREGO (40); 

*   CLASSE-CARGO(1); 

*   PADRÃO-CARGO(3); 

*   REFERENCIA-CARGO(2); 

*   NÍVEL_CARGO(3) ;

*   DESCRIÇÃO DO TIPO DE APOSENTADORIA (27); 

*   FUNDAMENTAÇÃO LEGAL DA APOSENTADORIA (40); 

*   ATO DE APOSENTADORIA(60); 

*   DATA DA APOSENTADORIA(DDMMAA); 

*   INGRESSO NO SERVIÇO PÚBLICO(50); 

*   DATA DE INGRESSO NO SERVIÇO PÚBLICO(DDMMAA) e VALOR MENSAL DA APOSENTADORIA(14)
